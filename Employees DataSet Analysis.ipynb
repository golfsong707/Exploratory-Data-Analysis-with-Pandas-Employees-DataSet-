## 5. Which department has the highest average salary?

```python
employee_data_df[['Department', 'Salary']]
```

```python
# i want to know which departments has the highest salary in the dept, and then compare with the department 
# that has the calculate the average salary within department.

df = employee_data_df[['Department', 'Salary']]

Dept_higest_salaries = df.groupby('Department')['Salary'].max()

print(Dept_higest_salaries)
```

```python
# from the  follwong departments, engineering Dept, was shown to have the highest salary among the employee, with is 149,794
```

```python
# but we like to see the which department, has th average highest salary, we do this 
# by grouping the by Department and salary culumn of the emplyee dataframe, and take the mean avarage.
```

```python
df = employee_data_df[['Department', 'Salary']]
```

```python
Dept_average_salaries = df.groupby('Department')['Salary'].mean()
```

```python
department_highest_average_salary = Dept_average_salaries.idxmax()
```

```python
print("Department with the highest average salary is:", department_highest_average_salary)
```
